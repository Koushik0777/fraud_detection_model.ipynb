```python
df.drop("Amount",axis=1,inplace=True)
```

```python
df.head()
```

```python
Target=df['Class']
Feature=df.drop('Class',axis=1)
```

```python
from sklearn.feature_selection import f_classif
f_scores, p_values = f_classif(Feature,Target)
f_df=pd.DataFrame({"Feature":Feature.columns,"F_scores":f_scores})


```

```python
f_df = f_df.sort_values(by='F_scores', ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(data=f_df.tail(15), x='F_scores', y='Feature')
plt.title("Top 15 Features by F-score")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(Feature,Target,test_size=0.2, random_state=42,stratify=Target)
print(Y_train.value_counts())
print(Y_test.value_counts())
```

```python
from sklearn.svm import SVC
```

```python
model=SVC(kernel='rbf',C=1)
```

```python
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_pred,Y_test)
print(accuracy)
```

```python
from sklearn.tree import DecisionTreeClassifier
dec_tree=DecisionTreeClassifier(class_weight="balanced",criterion='gini',splitter='best',max_depth=5,min_samples_split=20,min_samples_leaf=10,random_state=42
)
```

```python
dec_tree.fit(X_train,Y_train)
```

```python
Y_pred1=dec_tree.predict(X_test)
accuracy=accuracy_score(Y_test,Y_pred1)
print(accuracy)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
y_prob_dt = dec_tree.predict_proba(X_test)[:, 1]
print(classification_report(Y_test, Y_pred1))
print("ROC-AUC:", roc_auc_score(Y_test, y_prob_dt))
```
